## 개봉주차(wkNum)

```python
from datetime import datetime
```

```python
#연도를 주차별로 환산하는 함수를 만들어주자. 가장 오래된 연도가 2002년이므로 2000년을 기준으로 처리한다.
#15년, 9년, 4년은 53주차까지 있는 주 이므로 반영하여 환산해준다.
def yearToweek(yr):
    yr = yr-2000
    if(yr>15):
        return (yr-1)*52+3
    elif(yr>9):
        return (yr-1)*52+2
    elif(yr>4):
        return (yr-1)*52+1
    else:
        return (yr-1)*52
```

```python
#연도를 주차로 환산할 때에는 isocalendar를 사용해야한다. 따라서 isoYr, isoWk이라는 column을 만들어줘야한다.
isoYr = []
isoWk = []
isoDay = [] #요일을 의미. 1:월요일 2:화요일 ...

openYr = movie_total.openYr.tolist()
openMth = movie_total.openMth.tolist()
openDay = movie_total.openDay.tolist()

for i in movie_total.index.tolist():
    dt = datetime(openYr[i], openMth[i], openDay[i])
    isoYr.append(dt.isocalendar()[0])
    isoWk.append(dt.isocalendar()[1])
    isoDay.append(dt.isocalendar()[2])
    
wkNum = []

for i in movie_total.index.tolist():
    tempWk = yearToweek(isoYr[i]) + isoWk[i]
    wkNum.append(tempWk)

movie_total['wkDay'] = pd.Series(isoDay)
movie_total['wkNum'] = pd.Series(wkNum)
```

## 영화코드(movieCd)

이 영화코드(movieCd)는 영화진흥위원회(kobis)에서 영화를 식별하는 데 사용하는 코드이다.  
따라서 API를 사용하기 위해 json과 urllib를 import 해준다.

```python
import json
import urllib
```

```python
#YYYY-MM-DD의 구조인 openDt를 YYYYMMDD로 바꾸어주는 함수를 지정한다.
def dateConvert(date):
    tempDt = str(date)
    tempDt = tempDt[0:4]+tempDt[5:7]+tempDt[8:10]
    return tempDt
```

movieNm, openStartDt, openEndDt를 통해 movielist를 검색하면

만약 page가 변하지 않아서 곤란하다면 curPage와 itemPerPage를 조정해준다.

```python
#영화진흥원 API를 사용하기 위해 필요한 key들이다.
kobisKey = ['036ec0b9bd53529ea2dc74ea008e9d91', 'e4cca38919fb306d822c32d7d14d5a52', '20bfdc740feb247ca6eca9a04f5f9e86'
            ,'70e6e54392ec981956fb99e7e2f57e16', '82a0463123d6111d50a52206ecc82d69', '44e32445d382c479dfbfe7164e586109'
            ,'e6579370845ffc3095aeb29a301f2f37', 'ef1d12d352ac16d8509140f88ffd9902', '34ceb556cdfbeb005650d6cbaa13d5fe']
```

### 영화코드 ver1(탐색을 철저히)

```python
movieCd = []

for i in movie_total.index.tolist(): #movie_total의 index로 for문을 돌려준다.
    #조회할 movieNm
    tempNm = str(movieNm[i])
    #utf로 검색해야하기에 이렇게 만들어준다.
    tempNm = urllib.parse.quote(tempNm, safe='')
    
    #비교할 openDt
    tempDt = dateConvert(openDt[i])
    #조회할 openYr
    tempYr = str(openYr[i])
    
    #넣어줄 movieCd, movieCd가 채워질 때 까지 while문을 돌릴 것이므로, 빈 list를 만들어준다.
    tempCd = []
    #page수 초기화
    pageNum = 0
    
    #tempCd가 채워질 때까지 돌아가는 코드 작성
    while (len(tempCd)==0):
        #page수 추가
        pageNum = pageNum+1
        
        movieURL = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json?key={0}&curPage={1}&itemPerPage=100&movieNm={2}&openStartDt={3}&openEndDt={4}".format(kobisKey[0], pageNum, tempNm, tempYr, tempYr)
        moviePage = urllib.request.urlopen(movieURL)
        movieData = json.loads(moviePage.read())
        
        #찾는 결과가 하나일 경우, 이름과 개봉일자를 확인하고 입력
        if(movieData['movieListResult']['totCnt']==1):
            tempList = movieData['movieListResult']['movieList'][0]
            if ((tempList['movieNm']==str(movieNm[i]))&(tempList['openDt']==tempDt)):
                tempCd = [tempList['movieCd']]
        
        #찾는 결과가 0일 경우, 00000000 을 넣고 탈출
        elif(movieData['movieListResult']['totCnt']==0):
            tempCd = ['00000000']
            
        else:
            for j in movieData['movieListResult']['movieList']:
                if ((j['movieNm']==str(movieNm[i]))&(j['openDt']==tempDt)):
                    tempCd = [j['movieCd']]
                    break #for문을 나가는 break
                    
    movieCd = movieCd + tempCd
```

### 주의사항

## 등급(watchGrade)

전체 관람가, 12세 관람가, 15세 관람가, 청소년 관람불가 로 통일시켜준다.

```python
movie_total.watchGradeNm.drop_duplicates()
```

```python
def gradeConvert(grade):
    if ((grade=="15세이상관람가")|(grade=="15세관람가")|(grade=="15세관람가,15세이상관람가")):
        return "15세 관람가"
    elif ((grade=="12세이상관람가")|(grade=="12세관람가")|(grade=="12세이상관람가,12세관람가")):
        return "12세 관람가"
    elif ((grade=="청소년관람불가")|(grade=="18세관람가")):
        return "청소년 관람불가"
    elif (grade=="전체관람가"):
        return "전체 관람가"
    else:
        return grade
```

```python
watchGradeNm = movie_total.watchGradeNm
watchGradeNm = watchGradeNm.apply(gradeConvert)
movie_total.watchGradeNm = watchGradeNm
```
